##1. Read the dataset to the python environment.

```python

import pandas as pd
import matplotlib.pyplot as plt

cars_data = pd.read_csv("/content/cars_data.csv")
cars_data
```

##2. Check for the null values present in the dataset.

```python
cars_data.isnull().sum()
```

##3. Plot a bar graph of male vs female buyers participated in the sales.


```python
gender = cars_data["Car Gender"].value_counts(dropna=False)
gender
gender.plot(kind='bar', color = "blue")
plt.title("BAR GRAPH OF MALE VS FEMALE BUYERS")
plt.xlabel("BUYERS")
plt.ylabel("FREQUENCY")
```

## 4. Find the top 5 cars based on their sales price

```python
cars_data.nlargest(5, ['Sale Price'])
```

##5. Find the least 5 cars based on their Resell price

```python
cars_data.nsmallest(5, ['Resell Price'])
```
